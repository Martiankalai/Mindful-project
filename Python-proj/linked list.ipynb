```python
class Node:
    def __init__(self, value=None):
        self.value = value
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None
    
    def insert_at_beginning(self, value):
        new_node = Node(value)
        if not self.head:
            self.head = new_node
            return
        new_node.next = self.head
        self.head = new_node
    
    def insert_at_end(self, value):
        new_node = Node(value)
        if not self.head:
            self.head = new_node
            return
        
        last_node = self.head
        while last_node.next:
            last_node = last_node.next
        
        last_node.next = new_node
        
    def insert_after_element(self,value,pos_value):
        
        new_node=Node(value)
        temp=self.head
        while temp:
            if temp.value==pos_value:
                temp_node=temp.next
                temp.next=new_node
                new_node.next=temp_node
                return
            temp=temp.next 
        return
    
           
    def search_element(self,value):
        pos=0
        temp=self.head
        while temp.next:
            if value==temp.value:
               return pos
            pos+=1
            temp=temp.next
        return -1
        
    
    def deletion(self,value):
        cur=self.head
        prev=self.head
        while cur is not None:
            if cur.value==value:
                 prev.next=cur.next
                 return
            prev=cur
            cur=cur.next
        
       
    
    def display(self):
        current = self.head
        while current:
            print(current.value, end=" -> ")
            current = current.next
        print("None")
       
        
    def inser_end(self,value):
        new_node=Node(value)
        if self.head is None:
            self.head=new_node
        else:
            temp=self.head
            while temp.next!=None:
               temp=temp.next
            temp.next=new_node  



if __name__ == "__main__":
    
    linked_list = LinkedList()
    
    
    linked_list.insert_at_beginning(5)
    linked_list.insert_at_beginning(10)
    linked_list.insert_at_beginning(15)
    
    
    print("Linked List after inserting at the beginning:")
    linked_list.display()
    
    
    linked_list.insert_at_end(20)
    linked_list.insert_at_end(25)
    
    
    print("Linked List after inserting at the end:")
    linked_list.display()
    
   
    print("list the search element:", linked_list.search_element(10))
    
    linked_list.deletion(5)
    print("list the deleted element")
    linked_list.display()
    
    linked_list.insert_after_element(30,15)
    print("inserted element")
    linked_list.display()
    
    linked_list.inser_end(34)
    print("New")
    linked_list.display()
    
    
    
    
    
```
