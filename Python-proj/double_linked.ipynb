```python
class Node:
    def __init__ (self,value):
        self.value=value
        self.prev=None
        self.next=None
class doubly_linked:
    def __init__(self):
        self.head = None
        
    def insert_at_beginning(self,value):
        new_node=Node(value)
        if self.head is None:
            self.head=new_node
        else:
            temp=self.head
            new_node.next=temp
            self.head.prev=new_node
            self.head=new_node
            
    def insert_at_end(self,value):
        new_node=Node(value)
        if self.head is None:
            self.head=new_node
        else:
            temp=self.head
            while(temp.next!=None):
                temp=temp.next
            temp.next=new_node
            new_node.prev=temp
            
    def insert_after_ele(self,value,pos_value):
        new_node=Node(value)
        if self.head is None:
            self.head=new_node
        else:
            temp=self.head
            while(temp!=None):
                if(temp.value==pos_value):
                    new_node.next=temp.next
                    temp.next.prev=new_node
                    temp.next=new_node
                temp=temp.next
                
    def deletion(self,value):
        if (self.head.value==value):
            self.head.next.prev=None
            self.head=self.head.next
        else:
            temp=self.head
            while(temp!=None):
                if temp.value==value:
                    temp.prev.next=temp.next
                    temp.next.prev=temp.prev
                temp=temp.next
                
    def insert_before(self,value,pos_value):
        new_node=Node(value)
        if self.head is None:
            self.head=new_node
        temp=self.head
        while(temp!=None):
            if temp.value==pos_value:
                if temp.prev is None:
                    temp.prev=new_node
                    new_node.next=temp
                    self.head=new_node
                else:
                    new_node.prev=temp.prev    
                    temp.prev.next=new_node    
                    new_node.next=temp            
                    temp.prev=new_node
                return
            temp=temp.next
            
                
                    
                               
    def display(self):
        current = self.head
        while current:
            print(current.value, end=" ")
            current = current.next
        print()
       
dl=doubly_linked()
dl.insert_at_beginning(5)
dl.insert_at_beginning(10)
dl.insert_at_end(6)
dl.insert_at_end(7)
dl.insert_after_ele(15,10)
dl.display()
dl.deletion(6)
dl.display()
dl.insert_before(1,15)
dl.display()
            
```
